```python
from txtai.embeddings import Embeddings

emb = Embeddings(
    {
        "path": "sentence-transformers/all-MiniLM-L6-v2",
        "content": True
    }
)
```

```python
data = [
    {"id": 0, "text": "Machine learning is fun."},
    {"id": 1, "text": "Sejong University OSS project uses vector DB."}
]

emb.index(data)
```

```python
emb.search("vector database", 5)
```
